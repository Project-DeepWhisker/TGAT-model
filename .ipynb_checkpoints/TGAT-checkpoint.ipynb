**TGAT 網路入侵偵測專案 - 開發地圖**

**終極目標：** 開發一個高效能、高準確度的時間圖注意力網路模型，能夠處理大規模網路流量資料（如 CICDDoS 或真實封包），並有效偵測攻擊行為。

**階段一：基礎建設與效能瓶頸突破 (Foundational Improvements & Bottleneck Resolution)**

* **目標：** 建立穩定、可運行的開發環境，解決目前最嚴重的效能瓶頸（鄰居查找與批次處理），讓模型能以合理效率運行並真正利用圖結構。
* **主要任務：**
    1.  **環境與依賴項穩定化 (Task 1.1):**
        * **強烈建議：** 使用虛擬環境 (`venv` 或 `conda`) 隔離專案依賴。
        * 確保安裝**兼容**的 PyTorch (2.1.1+cu121)、PyG (e.g., 2.6.1) 及 PyG Temporal (建議升級至包含 `TemporalNeighborLoader` 的較新穩定版) [cite: 1, 260, 405]。
        * **必須：** 使用 PyG 官方指令安裝 `torch-scatter`, `torch-sparse` 等核心依賴項，確保與 PyTorch/CUDA 版本匹配。
        * 驗證 `import torch_geometric_temporal` 以及其 Loader（如 `TemporalNeighborLoader`）可以成功執行。
    2.  **實現高效資料載入與鄰居採樣 (Task 1.2):**
        * **核心替換：** 將訓練迴圈 (Cell 10) [cite: 1, 140, 303, 440] 中的**標準 PyTorch DataLoader 回退方案移除**。
        * **導入 PyGT Loader:** 學習並使用 `torch_geometric_temporal` 的 `TemporalNeighborLoader`（或其他如 `TemporalDataLoader`，但前者更適合 TGAT）。
        * **配置 Loader:** 設定合理的 `batch_size`、鄰居採樣策略與數量 (`num_neighbors` 參數，例如 `[10]` 或 `[15, 10]` 代表採樣一或兩跳鄰居)。
        * **準備 Dataset:** 確保您在 Cell 7 [cite: 1, 78, 243] 創建的 `TemporalData` 物件（或其底層數據）能被所選的 Loader 正確處理。
    3.  **適配模型 Forward Pass (Task 1.3):**
        * **修改 `TGAT.forward` (Cell 9)[cite: 1, 93, 286]:** 將其改回我們之前討論的「專業」版本，使其能接收並處理由 `TemporalNeighborLoader` 產生的 `batch` 物件（包含 `n_id`, `edge_index`, `e_id`, `size`, `edge_msg`, `edge_t` 等屬性）。
        * **移除內部迴圈：** 完全移除 `TGAT.forward` 內的逐事件 Python 迴圈和對 `find_k_recent_temporal_neighbors` 的呼叫。
        * **適配 Attention Layer:** 確保 `TemporalGraphAttentionLayer` (Cell 9) [cite: 1, 82, 265] 的 `forward` 方法能正確接收來自 `TGAT.forward` 傳遞的批次化參數（`x`, `edge_index`, `edge_t`, `edge_msg`, `size`）。
    4.  **實現基本狀態更新 (Task 1.4 - 可選但建議):**
        * 在訓練迴圈 `optimizer.step()` 之後，添加邏輯來更新模型的 `node_states` 和 `last_update` 緩衝區（參考 TGN 概念，使用模型輸出或中間狀態更新 `batch.n_id` 對應的節點）。

* **預期成果：** 訓練迴圈能以顯著提高的效率運行，模型真正開始利用鄰居資訊計算注意力，為後續準確度提升打下基礎。

**階段二：準確度提升與表示學習優化 (Accuracy Enhancement & Representation Learning)**

* **目標：** 在解決效能瓶頸後，專注於提升模型的偵測準確度，特別是針對攻擊類別的召回率 (Recall)。
* **主要任務：**
    1.  **節點定義精煉 (Task 2.1):**
        * **系統性實驗：** 嘗試更多不同的特徵組合來定義節點 ID [cite: 1, 28, 40, 211]。例如：(來源IP, 目標IP, 目標埠, 協議) - 如果能從 CICDDoS 或真實流量中提取 IP/埠號；或基於 `service`, `flag`, `protocol_type`, `dst_host_*` 特徵的組合。
        * **分析影響：** 評估不同定義方式產生的節點數量、圖的稀疏性，以及其對最終模型性能（Accuracy, Precision, Recall, F1）的影響。目標是找到能最好區分正常/攻擊互動模式的節點表示粒度。
    2.  **特徵工程與選擇 (Task 2.2):**
        * **重要性分析：** 評估當前 117 維特徵 [cite: 1, 57, 75, 240] 的重要性（例如使用模型本身的梯度或 SHAP 值）。
        * **降維/篩選：** 考慮是否可以移除冗餘或貢獻不大的特徵，或使用 PCA 等技術降維。
        * **新特徵構建：** 思考是否能從原始流量數據（特別是真實封包）中提取更有用的特徵，例如時間窗口內的統計特徵、包長序列、TLS/應用層資訊等（這可能需要在 GNN 外部進行）。
    3.  **模型架構與超參數調優 (Task 2.3):**
        * **TGAT 內部：** 調整 `n_layers`, `n_heads`, `time_dim`, `msg_dim`, `dropout` [cite: 1, 93, 286]。
        * **時間編碼：** 實驗不同的時間編碼方式（如 Time2Vec）。
        * **注意力機制：** 嘗試其他圖注意力變體。
        * **TGN 記憶體 (進階)：** 考慮用 GRU 或 RNN 取代簡單的 `node_states` buffer 來實現更複雜的節點記憶體更新。
    4.  **處理類別不平衡 (Task 2.4):**
        * **分析目標資料集：** 檢查 CICDDoS 或您的目標資料集的類別分佈。
        * **損失函數加權：** 如果不平衡，在 `BCEWithLogitsLoss` 中設定 `pos_weight`。
        * **評估指標側重：** 更關注 AUC-PR、Recall、F1 等對不平衡數據更敏感的指標。

* **預期成果：** 模型的整體準確度，特別是 Recall 指標得到顯著提升，模型能更好地泛化到未見過的攻擊模式。

**階段三：面向大規模數據與實際應用的擴展 (Scalability & Real-world Application)**

* **目標：** 確保優化後的方案能夠高效處理大規模資料集，並為部署到實際網路環境進行攻擊偵測做好準備。
* **主要任務：**
    1.  **大規模資料集測試 (Task 3.1):**
        * **數據處理：** 開發適用於 CICDDoS 等大型資料集的數據讀取和前處理流程（可能需要分塊處理或使用更高效的資料庫/工具）。
        * **效能標竿：** 在 CICDDoS 上測試使用 `TemporalNeighborLoader` 的訓練速度和記憶體消耗，優化 Loader 參數（`batch_size`, `num_neighbors`, `num_workers` 等）。
    2.  **性能分析與優化 (Task 3.2):**
        * **剖析瓶頸：** 使用 PyTorch Profiler 等工具分析訓練和推論過程中的 GPU/CPU 瓶頸。
        * **進階優化：** 探索混合精度訓練 (Mixed Precision) 等技術進一步加速。
    3.  **串流處理與線上偵測 (Task 3.3):**
        * **架構設計：** 設計能處理連續網路封包流或日誌流的推論 (Inference) 管線。
        * **模型適應：** 可能需要調整模型或狀態更新機制以適應線上、持續學習的場景（如果需要模型隨時間演進）。
    4.  **部署考量 (Task 3.4):**
        * **模型序列化：** 將訓練好的模型轉換為易於部署的格式（如 TorchScript, ONNX）。
        * **服務化：** 使用 Flask/FastAPI 封裝模型提供 API 接口，或使用 TorchServe 等專用工具進行部署。
        * **整合：** 考慮如何與現有的網路監控系統（如 SIEM, IDS）整合。
    5.  **攻擊模擬環境驗證 (Task 3.5 - 可選):**
        * **整合 Caldera[cite: 2, 183]:** 將 Caldera 生成的攻擊流量或日誌輸入到您的偵測模型中，評估模型在模擬真實攻擊場景下的表現。

* **預期成果：** 形成一套可擴展、高效、準確的時間圖網路入侵偵測解決方案，具備部署到實際環境的潛力。

```python
import pandas as pd
import numpy as np

# 定義 NSL-KDD 的 41 個特徵名稱 + 攻擊標籤 + 難度標籤
col_names = ["duration","protocol_type","service","flag","src_bytes",
             "dst_bytes","land","wrong_fragment","urgent","hot","num_failed_logins",
             "logged_in","num_compromised","root_shell","su_attempted","num_root",
             "num_file_creations","num_shells","num_access_files","num_outbound_cmds",
             "is_host_login","is_guest_login","count","srv_count","serror_rate",
             "srv_serror_rate","rerror_rate","srv_rerror_rate","same_srv_rate",
             "diff_srv_rate","srv_diff_host_rate","dst_host_count","dst_host_srv_count",
             "dst_host_same_srv_rate","dst_host_diff_srv_rate","dst_host_same_src_port_rate",
             "dst_host_srv_diff_host_rate","dst_host_serror_rate","dst_host_srv_serror_rate",
             "dst_host_rerror_rate","dst_host_srv_rerror_rate","attack", "difficulty"]

# 載入 KDDTrain+_20Percent.txt 和 KDDTest+.txt
train_path = './KDDTrain+_20Percent.txt'
test_path = './KDDTest+.txt'

# *** 修改：將原始資料載入 df_train_orig 和 df_test_orig ***
df_train_orig = pd.read_csv(train_path, header=None, names=col_names)
df_test_orig = pd.read_csv(test_path, header=None, names=col_names)

print("原始訓練集樣本數:", df_train_orig.shape)
print("原始測試集樣本數:", df_test_orig.shape)
print("原始訓練集特徵數 (含標籤):", df_train_orig.shape[1])

# 顯示前幾行數據
print("\n原始訓練集前 5 行:")
print(df_train_orig.head())
```

```python
import torch # 確保導入 torch
import pandas as pd # 確保導入 pandas

print("--- 開始定義節點 ID ---")
# 使用原始 DataFrame (df_train_orig, df_test_orig)
# 再次強調：'protocol_type' 和 'service' 僅為範例，請選擇有意義的欄位
# 您可以根據需要調整或選擇不同的欄位組合
try:
    df_train_orig['src_entity'] = df_train_orig['protocol_type'].astype(str) + "_" + df_train_orig['src_bytes'].astype(str) # 範例
    df_train_orig['dst_entity'] = df_train_orig['service'].astype(str) + "_" + df_train_orig['dst_bytes'].astype(str)     # 範例
    df_test_orig['src_entity'] = df_test_orig['protocol_type'].astype(str) + "_" + df_test_orig['src_bytes'].astype(str)
    df_test_orig['dst_entity'] = df_test_orig['service'].astype(str) + "_" + df_test_orig['dst_bytes'].astype(str)

    # 合併所有可能的實體來創建節點映射
    all_entities = pd.concat([df_train_orig['src_entity'], df_train_orig['dst_entity'],
                               df_test_orig['src_entity'], df_test_orig['dst_entity']]).unique()
    entity_to_id = {entity: i for i, entity in enumerate(all_entities)}
    num_nodes = len(all_entities) # 全局的節點數量
    print(f"基於定義的唯一節點數量: {num_nodes}")

    # 將實體標識符映射到整數節點 ID 並儲存
    train_src_ids = torch.tensor(df_train_orig['src_entity'].map(entity_to_id).values, dtype=torch.long)
    train_dst_ids = torch.tensor(df_train_orig['dst_entity'].map(entity_to_id).values, dtype=torch.long)
    test_src_ids = torch.tensor(df_test_orig['src_entity'].map(entity_to_id).values, dtype=torch.long)
    test_dst_ids = torch.tensor(df_test_orig['dst_entity'].map(entity_to_id).values, dtype=torch.long)

    # 移除輔助欄位
    df_train_orig = df_train_orig.drop(columns=['src_entity', 'dst_entity'])
    df_test_orig = df_test_orig.drop(columns=['src_entity', 'dst_entity'])
    print("--- 節點 ID 定義和映射完成 ---")

except KeyError as e:
    print(f"發生 KeyError: {e}。請確保 df_train_orig 和 df_test_orig 已正確載入且包含所需欄位。")
except Exception as e:
    print(f"定義節點 ID 時發生錯誤: {e}")
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
import numpy as np # 確保導入 numpy
import pandas as pd # 確保導入 pandas

print("--- 開始特徵前處理 (OHE + Scaling) ---")

try:
    # 1. 分離標籤和特徵
    train_labels = df_train_processed['binary_attack']
    test_labels = df_test_processed['binary_attack']
    df_train_features = df_train_processed.drop(columns=['binary_attack'])
    df_test_features = df_test_processed.drop(columns=['binary_attack'])

    # 2. 識別特徵類型
    categorical_features = ['protocol_type', 'service', 'flag']
    # 確保數值特徵列表從當前的特徵 DataFrame 中獲取
    numerical_features = df_train_features.select_dtypes(include=np.number).columns.tolist()

    # 移除恆為 0 的 num_outbound_cmds (如果還在)
    if 'num_outbound_cmds' in df_train_features.columns:
        if 'num_outbound_cmds' in numerical_features:
             numerical_features.remove('num_outbound_cmds')
        df_train_features = df_train_features.drop(['num_outbound_cmds'], axis=1)
        df_test_features = df_test_features.drop(['num_outbound_cmds'], axis=1)
    else:
        # 如果之前移除了，確保它不在 numerical_features 列表中
        if 'num_outbound_cmds' in numerical_features:
             numerical_features.remove('num_outbound_cmds')


    # 3. 創建 ColumnTransformer Pipeline (包含 OHE 和 Scaler)
    preprocessor = ColumnTransformer(
        transformers=[
            ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_features),
            ('num', StandardScaler(), numerical_features) # 對數值特徵進行縮放
        ],
        remainder='drop' # 確保只處理指定的特徵
    )

    # 4. 擬合訓練集特徵並轉換訓練集和測試集
    print("擬合 Preprocessor...")
    preprocessor.fit(df_train_features)
    print("轉換訓練集...")
    train_features_processed_array = preprocessor.transform(df_train_features)
    print("轉換測試集...")
    test_features_processed_array = preprocessor.transform(df_test_features)

    # 5. 獲取最終的特徵名稱
    ohe_feature_names = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_features)
    final_feature_names = list(ohe_feature_names) + numerical_features
    print(f"最終用於模型的特徵數量: {len(final_feature_names)}")

    # 6. 轉換回 DataFrame (用於創建 TemporalData)
    df_train_final_features = pd.DataFrame(train_features_processed_array, columns=final_feature_names, index=df_train_features.index)
    df_test_final_features = pd.DataFrame(test_features_processed_array, columns=final_feature_names, index=df_test_features.index)

    print(f"\n最終處理後訓練集特徵維度: {df_train_final_features.shape}")
    print(f"最終處理後測試集特徵維度: {df_test_final_features.shape}")
    print("\n處理後訓練集數值特徵 (部分) 的前 5 行:")
    # 顯示縮放後的數值特徵（如果需要檢查）
    print(df_train_final_features[numerical_features].head())

except KeyError as e:
     print(f"特徵前處理時發生 KeyError: {e}. 請檢查欄位名稱和 DataFrame 的狀態。")
except Exception as e:
    print(f"特徵前處理時發生錯誤: {e}")
print("--- 特徵前處理完成 ---")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn.inits import glorot, zeros
from torch_geometric.data import TemporalData # 需要導入 TemporalData
import numpy as np
try:
    from torch_scatter import scatter_add, scatter_softmax
    _torch_scatter_available = True
except ImportError:
    _torch_scatter_available = False
    print("警告：torch_scatter 未安裝，TemporalGraphAttentionLayer 將無法正確執行注意力機制。")

# 1. Functional Time Encoding (FTE) 模組 (與之前相同)
class FunctionalTimeEncoder(nn.Module):
    def __init__(self, d_model):
        super(FunctionalTimeEncoder, self).__init__()
        self.d_model = d_model
        self.omega = nn.Parameter(torch.randn(d_model // 2))

    def forward(self, t):
        if t.ndim == 1:
            t = t.unsqueeze(-1)
        t = t.float()
        cos_t = torch.cos(self.omega * t)
        sin_t = torch.sin(self.omega * t)
        phi_t = torch.cat([cos_t, sin_t], dim=-1) / torch.sqrt(torch.tensor(self.d_model / 2.0, device=t.device))
        return phi_t

# 2. Multi-Head Temporal Graph Attention Layer (保持使用 scatter 的版本，但注意輸入來源)
class TemporalGraphAttentionLayer(nn.Module):
    # ... (init 部分與上次提供的 v0.54.0 兼容版本相同) ...
    def __init__(self, in_channels, msg_channels, time_channels, out_channels, n_heads=2, dropout=0.1):
        super(TemporalGraphAttentionLayer, self).__init__()
        self.in_channels = in_channels
        self.msg_channels = msg_channels
        self.time_channels = time_channels
        self.out_channels = out_channels
        self.n_heads = n_heads
        self.dropout = dropout

        self.time_encoder = FunctionalTimeEncoder(time_channels)

        combined_dim_kv = in_channels + msg_channels + time_channels
        combined_dim_q = in_channels + msg_channels + time_channels

        self.W_q = nn.Linear(combined_dim_q, out_channels * n_heads)
        self.W_k = nn.Linear(combined_dim_kv, out_channels * n_heads)
        self.W_v = nn.Linear(combined_dim_kv, out_channels * n_heads)

        self.W_o = nn.Linear(out_channels * n_heads, out_channels)
        self.ffn = nn.Sequential(
            nn.Linear(out_channels + in_channels, out_channels),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(out_channels, out_channels)
        )
        self.layer_norm_agg = nn.LayerNorm(out_channels * n_heads)
        self.layer_norm_out = nn.LayerNorm(out_channels)
        self.attn_dropout = nn.Dropout(dropout)

    def forward(self, x, last_update_t, neighbors_info):
        """
        這個 forward 方法現在由 TGAT 內部的迴圈呼叫，
        處理單個目標節點及其找到的鄰居 (如果 neighbors_info 非 None)。
        x: [1, in_channels]
        last_update_t: [1]
        neighbors_info: dict or None
        """
        batch_size = x.size(0) # 這裡 batch_size 應該是 1

        # --- 準備 Q ---
        target_time_enc = self.time_encoder(torch.zeros(batch_size, 1, device=x.device))
        zero_msg = torch.zeros(batch_size, self.msg_channels, device=x.device)
        target_feat_for_q = torch.cat([x, zero_msg, target_time_enc], dim=-1)
        q = self.W_q(target_feat_for_q).view(batch_size, self.n_heads, self.out_channels)

        if not _torch_scatter_available:
             # 如果沒有 scatter，無法執行注意力，直接返回 FFN 結果
             h_agg_projected = torch.zeros(batch_size, self.out_channels, device=x.device)
             combined_input = torch.cat([h_agg_projected, x], dim=-1)
             output = self.ffn(combined_input)
             output = self.layer_norm_out(output)
             return output

        if not neighbors_info or neighbors_info.get('num_neighbors', 0) == 0:
            # --- 無鄰居情況 ---
            h_agg = torch.zeros(batch_size, self.n_heads * self.out_channels, device=x.device)
        else:
            # --- 有鄰居情況 ---
            neighbors_x = neighbors_info['neighbors_x']       # [N_neigh, in_channels]
            neighbors_t = neighbors_info['neighbors_t']       # [N_neigh]
            neighbors_msg = neighbors_info['neighbors_msg']     # [N_neigh, msg_channels]
            current_t = neighbors_info['current_t']           # scalar or tensor[1]
            # neighbor_indices 在單節點處理時都是 0
            neighbor_indices = torch.zeros(neighbors_info['num_neighbors'], dtype=torch.long, device=x.device)
            num_neighbors = neighbors_info['num_neighbors']

            # 時間差編碼
            if isinstance(current_t, torch.Tensor):
                 current_t_expanded = current_t[neighbor_indices] # 都是 0，所以取第一個
            else: # python 數字
                 current_t_expanded = torch.tensor(current_t, device=x.device) # 保持標量

            delta_t = current_t_expanded - neighbors_t
            neighbor_time_enc = self.time_encoder(delta_t)

            # 準備 K, V
            neighbor_feat_for_kv = torch.cat([neighbors_x, neighbors_msg, neighbor_time_enc], dim=-1)
            k = self.W_k(neighbor_feat_for_kv).view(num_neighbors, self.n_heads, self.out_channels)
            v = self.W_v(neighbor_feat_for_kv).view(num_neighbors, self.n_heads, self.out_channels)

            # 擴展 Q
            q_expanded = q[neighbor_indices] # shape [N_neigh, H, C]

            # 縮放 Q
            q_scaled = q_expanded / torch.sqrt(torch.tensor(self.out_channels, dtype=torch.float, device=x.device))

            # 計算原始注意力分數
            attn_scores_raw = (q_scaled * k).sum(dim=-1) # [N_neigh, H]

            # Softmax (因為 target index 都是 0，等同於對所有鄰居做 softmax)
            alpha_weights = F.softmax(attn_scores_raw, dim=0) # [N_neigh, H]
            alpha_weights = self.attn_dropout(alpha_weights)

            # 聚合 Value
            alpha_weights_expanded = alpha_weights.unsqueeze(-1)
            v_weighted = v * alpha_weights_expanded

            # 聚合到目標節點 (因為只有一個目標節點，直接求和)
            h_agg_sum = v_weighted.sum(dim=0) # [H, C]
            h_agg = h_agg_sum.view(batch_size, -1) # [1, H*C]

        # --- 合併與 FFN ---
        h_agg = self.layer_norm_agg(h_agg)
        h_agg = F.dropout(h_agg, p=self.dropout, training=self.training)
        h_agg_projected = self.W_o(h_agg) # [1, C]

        combined_input = torch.cat([h_agg_projected, x], dim=-1) # [1, C + in_channels]
        output = self.ffn(combined_input)
        output = self.layer_norm_out(output)

        return output


# --- 鄰居查找輔助函式 (之前提供，低效版本) ---
def find_k_recent_temporal_neighbors(target_node_id, current_t, data_cpu, k, node_states, device):
    """ (之前提供的簡化且低效版本) """
    mask = (data_cpu.t < current_t) & ((data_cpu.src == target_node_id) | (data_cpu.dst == target_node_id))
    past_interaction_indices = mask.nonzero(as_tuple=False).squeeze(-1)
    if past_interaction_indices.numel() == 0: return None
    past_times = data_cpu.t[past_interaction_indices]
    sorted_time_indices = torch.argsort(past_times, descending=True)
    recent_indices = past_interaction_indices[sorted_time_indices[:k]]
    num_neighbors = len(recent_indices)
    if num_neighbors == 0: return None
    recent_src = data_cpu.src[recent_indices]
    recent_dst = data_cpu.dst[recent_indices]
    recent_t = data_cpu.t[recent_indices]
    recent_msg = data_cpu.msg[recent_indices]
    is_src_mask = (recent_src == target_node_id)
    neighbors_node_ids = torch.where(is_src_mask, recent_dst, recent_src)
    neighbors_node_ids_on_device = neighbors_node_ids.to(node_states.device)
    try:
        neighbors_x = node_states[neighbors_node_ids_on_device]
    except IndexError: return None # 索引錯誤處理
    neighbors_info = {
        'neighbors_node_ids': neighbors_node_ids.to(device), 'neighbors_x': neighbors_x.to(device),
        'neighbors_t': recent_t.to(device), 'neighbors_msg': recent_msg.to(device),
        'neighbor_indices': torch.zeros(num_neighbors, dtype=torch.long, device=device),
        'num_neighbors': num_neighbors
    }
    return neighbors_info


# 3. 完整的 TGAT 模型 (**forward 方法改回逐事件處理**)
class TGAT(nn.Module):
    def __init__(self, num_nodes, msg_dim, time_dim, output_dim, n_layers=2, n_heads=2, dropout=0.1):
        super(TGAT, self).__init__()
        if num_nodes <= 0: raise ValueError("num_nodes must be positive.")
        self.num_nodes = num_nodes
        self.msg_dim = msg_dim
        self.time_dim = time_dim
        self.output_dim = output_dim
        self.n_layers = n_layers

        # 節點狀態存儲 (使用 Buffer)
        self.register_buffer('node_states', torch.zeros(num_nodes, msg_dim))
        self.register_buffer('last_update', torch.zeros(num_nodes))

        # TGAT 層
        self.tgat_layers = nn.ModuleList()
        current_in_ch = msg_dim
        for i in range(n_layers):
            current_out_ch = msg_dim
            self.tgat_layers.append(
                 TemporalGraphAttentionLayer(current_in_ch, msg_dim, time_dim, current_out_ch, n_heads, dropout)
            )
            current_in_ch = current_out_ch

        # 最終輸出層
        self.output_layer = nn.Linear(current_in_ch, output_dim)

    def forward(self, data: TemporalData):
        """
        接收 TemporalData 物件 (或其切片)，並在內部進行逐事件處理。
        """
        src, dst, t, msg, y = data.src, data.dst, data.t, data.msg, data.y
        batch_size = len(src) # 批次中的事件數量

        # 使用 clone()
        current_node_states = self.node_states.clone()
        current_last_update = self.last_update.clone()

        # *** 將輸入數據移到 CPU 以便鄰居查找 (如果需要且不在 CPU) ***
        data_cpu = data.cpu()

        outputs = []
        k_neighbors = 10 # 示例鄰居數量

        # --- 迴圈處理批次中的每個事件 ---
        for i in range(batch_size):
            current_src_node = src[i].item()
            current_dst_node = dst[i].item()
            current_t_scalar = t[i].item()
            # current_msg_vector = msg[i] # 在鄰居查找中使用 data_cpu.msg[index]

            # 1. 獲取狀態 (在模型所在 device)
            #    確保索引不超界
            if current_dst_node >= self.num_nodes or current_src_node >= self.num_nodes:
                 print(f"警告：節點 ID {current_dst_node} 或 {current_src_node} 超出範圍 {self.num_nodes}，跳過事件 {i}")
                 # 可以附加一個預設輸出或直接跳過
                 outputs.append(torch.zeros(self.output_dim, device=self.output_layer.weight.device)) # 示例：添加零輸出
                 continue

            dst_state = current_node_states[current_dst_node]
            dst_last_update_tensor = current_last_update[current_dst_node].unsqueeze(0)

            # 2. **呼叫鄰居查找函式 (逐事件)**
            neighbors_info_dst = find_k_recent_temporal_neighbors(
                target_node_id=current_dst_node,
                current_t=current_t_scalar,
                data_cpu=data_cpu, # 使用 CPU 上的數據
                k=k_neighbors,
                node_states=current_node_states, # 傳遞當前狀態
                device=dst_state.device         # 輸出到模型所在 device
            )

            # 為 neighbors_info 添加當前時間
            if neighbors_info_dst is not None:
                neighbors_info_dst['current_t'] = torch.tensor([current_t_scalar], device=dst_state.device)


            # 3. 通過 TGAT 層 (處理單個目標節點)
            h = dst_state.unsqueeze(0) # [1, msg_dim]
            for layer in self.tgat_layers:
                h = layer(h, dst_last_update_tensor, neighbors_info_dst)

            # 4. 預測
            event_output = self.output_layer(h.squeeze(0))
            outputs.append(event_output)

            # 5. 狀態更新 (簡化 - 可選，且僅更新了 dst)
            # if h.shape[1] == self.msg_dim:
            #     current_node_states[current_dst_node] = h.squeeze(0).detach()
            #     current_last_update[current_dst_node] = torch.tensor(current_t_scalar, device=current_last_update.device) # 更新 Buffer
            # 同樣可能需要更新 src 狀態

        # --- 迴圈結束 ---

        # # 更新模型狀態 (如果選擇在 forward 內部更新)
        # self.node_states.data.copy_(current_node_states)
        # self.last_update.data.copy_(current_last_update)

        if not outputs:
            return torch.empty((0, self.output_dim), device=self.output_layer.weight.device)

        return torch.stack(outputs) # [batch_size_events, output_dim]
```

```python
import torch
import torch.optim as optim
import torch.nn as nn
from torch_geometric.data import TemporalData # 確保導入
import pandas as pd
import numpy as np
# 導入標準 Loader 和 Dataset
from torch.utils.data import DataLoader, Dataset
import traceback # 用於打印詳細錯誤

# --- 假設相關類別和變數已定義 ---
# TGAT (使用接收 TemporalData 的 forward 版本), train_data, test_data, num_nodes, etc.
# find_k_recent_temporal_neighbors 函數也假設已定義 (在 Cell 9)

print("--- 開始設定訓練環境 ---")

# 1. 檢查 GPU 可用性並設定裝置 (與之前相同)
if torch.cuda.is_available(): device = torch.device('cuda')
else: device = torch.device('cpu')
print(f"將使用裝置: {device}")

# 獲取數據維度 (與之前相同)
# ... (省略，假設 msg_dim, time_dim, output_dim, num_nodes 已正確獲取) ...
if 'msg_dim' not in locals(): raise ValueError("msg_dim 未定義")
if 'num_nodes' not in locals() or num_nodes <= 0: raise ValueError("num_nodes 未定義或無效")
print(f"使用的節點數: {num_nodes}")
time_dim = 64; output_dim = 1


# 2. 實例化模型並將其移動到指定裝置 (與之前相同)
try:
    # 確保 TGAT 類別已定義 (使用接收 TemporalData 的 forward 版本)
    model = TGAT(num_nodes=num_nodes, msg_dim=msg_dim, time_dim=time_dim, output_dim=output_dim, n_layers=2, n_heads=2, dropout=0.1)
    model.to(device)
    print(f"模型已成功移動到裝置: {next(model.parameters()).device}")
except NameError: print("錯誤：TGAT 未定義"); raise
except Exception as e: print(f"模型錯誤: {e}"); raise


# --- 3. 準備 DataLoader (返回數據組件 + 預設 Collate) ---
train_loader = None
test_loader = None

try:
    if 'train_data' not in locals() or train_data is None: raise NameError("train_data 未定義")

    # 保持數據在 CPU 以便 Dataset 訪問
    train_data_cpu = train_data.cpu()
    if 'test_data' in locals() and test_data is not None:
        test_data_cpu = test_data.cpu()
    else:
        test_data_cpu = None

    batch_size = 128 # 設定批次大小
    print(f"\n--- 使用標準 DataLoader (返回組件 + 預設 Collate, batch_size={batch_size}) ---")

    # Dataset 返回每個事件的數據組件
    class ComponentTemporalDataset(Dataset):
        def __init__(self, temporal_data):
            self.data = temporal_data
            if not hasattr(temporal_data, 'src') or self.data.src is None:
                 raise ValueError("TemporalData 物件無效或缺少 'src'")
            self.num_events = self.data.src.size(0)
            required_attrs = ['src', 'dst', 't', 'msg', 'y'] # 驗證必要屬性
            for attr in required_attrs:
                 if not hasattr(self.data, attr) or getattr(self.data, attr) is None: raise ValueError(f"TemporalData 缺少: {attr}")
                 if not isinstance(getattr(self.data, attr), torch.Tensor): raise TypeError(f"{attr} 不是 Tensor")
                 if getattr(self.data, attr).size(0) != self.num_events: raise ValueError(f"{attr} 長度不符")

        def __len__(self):
            return self.num_events

        def __getitem__(self, idx):
            return ( self.data.src[idx], self.data.dst[idx], self.data.t[idx],
                     self.data.msg[idx], self.data.y[idx] )

    train_dataset = ComponentTemporalDataset(train_data_cpu)
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=False) # 使用預設 collate

    if test_data_cpu is not None:
       test_dataset = ComponentTemporalDataset(test_data_cpu)
       # 注意：預設 collate 會處理好 test_dataset 的輸出
       test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
       print("測試集 DataLoader 已創建。")
    else:
       print("未創建測試集 DataLoader。")

    print("DataLoader 已創建 (返回批次化的組件張量)。")

except NameError as e:
    print(f"錯誤：創建 DataLoader 時缺少變數: {e}")
    raise
except Exception as e:
    print(f"創建 DataLoader 時發生錯誤: {e}")
    raise


# 定義損失函數和優化器
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 訓練迴圈 (添加 ground_truth 的裝置移動)
epochs = 10
print("\n--- 開始訓練 (批次重建 TemporalData + 手動移動標籤) ---")

if train_loader is not None:
    for epoch in range(epochs):
        model.train()
        total_loss = 0
        num_processed_events = 0
        # train_loader 現在 yield 一個包含批次張量的元組
        for batch_idx, batch_components in enumerate(train_loader):
            try:
                # 解包批次組件 (來自 CPU)
                src_b, dst_b, t_b, msg_b, y_b = batch_components

                # 從批次組件重建 TemporalData 物件 (仍在 CPU)
                batch_data = TemporalData(src=src_b, dst=dst_b, t=t_b, msg=msg_b, y=y_b)

                # 將重建的批次 TemporalData 移到 GPU
                batch_data = batch_data.to(device)

                optimizer.zero_grad()

                # 模型 forward (輸入在 GPU, 輸出在 GPU)
                out = model(batch_data)

                # 計算損失前的準備
                output_processed = out.squeeze(-1) # 在 GPU
                ground_truth = batch_data.y.float() # 從 GPU 上的 batch_data 取 y，應在 GPU

                # *** 新增/修改：確保 ground_truth 在正確的 device 上 ***
                ground_truth = ground_truth.to(device)

                # # 移除之前的診斷打印
                # print(f"--- Batch {batch_idx} ---")
                # print(f"Device of model output ('out'): {output_processed.device}")
                # print(f"Device of ground_truth ('y'): {ground_truth.device}")

                # 計算損失 (現在兩者都應在 GPU 上)
                loss = criterion(output_processed, ground_truth)

                loss.backward()
                optimizer.step()

                total_loss += loss.item() * batch_data.num_events
                num_processed_events += batch_data.num_events

            except Exception as e:
                print(f"\nEpoch {epoch+1}, 處理批次 {batch_idx} 時發生錯誤: {e}")
                traceback.print_exc()
                break # 中斷當前 epoch

        if num_processed_events > 0:
            avg_loss = total_loss / num_processed_events
            print(f'Epoch {epoch+1}/{epochs}, Average Loss: {avg_loss:.4f}')
        else:
            print(f"Epoch {epoch+1}/{epochs}, 沒有成功處理的批次。")

else:
    print("因無法創建 DataLoader，訓練迴圈已跳過。")

print("--- 訓練完成! ---")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd # 確保導入 pandas
import torch      # 確保導入 torch

print("\n--- 開始評估 ---")
# 確保模型和測試數據在正確的裝置上
# 假設 model 和 test_data 已移動到 device

try:
    model.eval() # 設置模型為評估模式
    with torch.no_grad(): # 禁用梯度計算
        # 使用測試數據進行預測
        test_out = model(test_data)
        test_pred_prob = torch.sigmoid(test_out).squeeze() # 應用 sigmoid 獲取概率
        test_pred_binary = (test_pred_prob > 0.5).float() # 使用 0.5 作為閾值

    # 獲取真實標籤和預測標籤 (移回 CPU)
    y_true = test_data.y.cpu().numpy()
    y_pred = test_pred_binary.cpu().numpy()

    # 計算評估指標
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, zero_division=0)
    recall = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)
    cm = confusion_matrix(y_true, y_pred)

    print("\n模型評估結果:")
    print(f"  Accuracy:  {accuracy:.4f}")
    print(f"  Precision: {precision:.4f}")
    print(f"  Recall:    {recall:.4f}")
    print(f"  F1-Score:  {f1:.4f}")

    # 繪製混淆矩陣
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Normal', 'Attack'], yticklabels=['Normal', 'Attack'])
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title('Confusion Matrix')
    plt.show()

    # 將結果整理到表格中
    results_summary = pd.DataFrame({
        'Metric': ['Accuracy', 'Precision', 'Recall', 'F1-Score'],
        'Score': [accuracy, precision, recall, f1]
    })
    print("\n性能指標總結:")
    print(results_summary)

except NameError as e:
    print(f"評估時發生 NameError: {e}. 請確保 model 和 test_data 已成功定義並移至裝置。")
except Exception as e:
    print(f"評估時發生錯誤: {e}")

print("--- 評估完成 ---")
```
